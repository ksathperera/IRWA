Create a new dictionary by extracting the following keys from a given dictionary
Given Dictionary
sampleDict = {
"name": "Kelly",
"age":25,
"salary": 8000,
"city": "New york"
}
Keys to extract
keys = ["name", "salary"]
Answer = {'name': 'Kelly', 'salary': 8000}

```python
#Given Dictionary
sampleDict = {
"name": "Kelly",
"age":25,
"salary": 8000,
"city": "New york"
}

#Keys to extract
keys = ["name", "salary"]
Answer = {'name': 'Kelly', 'salary': 8000}

answer = dict()
for i in sampleDict.keys() and keys:
  answer[i] = sampleDict[i]
print(answer)
```

Given a list iterate it and display numbers which are divisible by 5 and if you find number
greater than 150 stop the loop iteration
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]
Answer =
15
55

150

```python
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]

for i in list1:
  if i > 150:
    break
  if (i % 5) == 0:
    print(i)
```

Count all lower case, upper case, digits, and special symbols from a given string
input_str = "P@#yn26at^&i5ve"
Total counts of chars, digits,and symbols
Chars = 8
Digits = 3
Symbol = 4

```python
input_str = "P@#yn26at^&i5ve"

lcount=0;
upcount=0;
digcount=0;
symbol=0;

for char in input_str:
  if char.islower():
    lcount+=1
  elif char.isupper():
    upcount+=1
  elif char.isdigit():
    digcount+=1
  else:
    symbol+=1

print("lower count: ",lcount,"uppercase count: ",upcount,"digit count",digcount,"symbols count:",symbol)
```
