```python
import os
#import re to preprocessing part
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize  #to split words
from nltk.stem import PorterStemmer

for doc_no in range(1,4):

  # Reading the document: It reads the contents of the document file (doc_1.txt, doc_2.txt, etc.) using the open() function and stores the content in the variable s. The contents are read
  # as a single line and any newline characters are replaced with a space.

  with open(os.getcwd()+'/doc_' + str(doc_no) + '.txt', 'r') as file:
    s = file.read().replace('\n','')[0:]
    #print('sentances: ',s)

    #preprocessing
    #normalizing
    #replacing the unnessary characters( won't - > will not )
    # Normalizing the text: The code applies a series of regular expression substitutions to normalize the text. For example, it replaces contractions like "won't" with "will not", "can't" with "can not",
    # and so on. It also removes unnecessary characters and replaces commas with

    s = re.sub(' ',' ',s)
    s = re.sub(r"won't","will not",s)
    s = re.sub(r"can\'t","can not",s)
    s = re.sub(r"n\'t","not",s)
    s = re.sub(r"\'re","are",s)
    s = re.sub(r"\'s","is",s)
    s = re.sub(r"\'d","would",s)
    s = re.sub(r"\'ll","will",s)
    s = re.sub(r"\'t","not",s)
    s = re.sub(r"\'m","am",s)
    s = re.sub(r'[^\w\s]','',s)  #remove white spaces

    #print("sentences :",s)

    #to tokenize
    words = word_tokenize(str(s).lower())
    #print(words)

    #to remove stopwords in sentences
    stop_words = stopwords.words('english')
    new_sentence = [word for word in words if word not in stop_words]
    print(new_sentence)

```

```python
#to tokenize
import nltk

nltk.download('stopwords')
nltk.download('punkt')
```

```python
#to stemming using porterstemmer
ps = PorterStemmer()
stemmed=[]

for i in new_sentence:
  stemmed.append(ps.stem(i))

print(stemmed)
```

```python
#to make positional index one, we have to use stemmed list
#positonal index
dictionary = dict()
temp_dictionary = dict()  #This line initializes an empty dictionary called temp_dictionary that will be used to store the positional index.
a=0
for x in stemmed:
  key=x  #This line assigns the current word in the loop (x) to the variable key.

  #This line checks if the key exists in the temp_dictionary.
  #If it doesn't exist, it adds the key to the dictionary as a key and initializes an empty list as its value.
  #This ensures that each unique word in the stemmed list has an entry in the dictionary.

  temp_dictionary.setdefault(key,[])

  temp_dictionary[key].append(a)  #This line appends the current value of a to the list corresponding to the key in the temp_dictionary.
  a+=1

print(temp_dictionary)

for x in temp_dictionary:
  #check whether that word is there in the dictionary or not
  if dictionary.get(x):
    dictionary[x][doc_no] = temp_dictionary.get(x)
  else:
    key=x
    dictionary.setdefault(key,[])
    dictionary[key]={}
    dictionary[x][doc_no]=temp_dictionary.get(x)


```
