```python
#jc = (x n y) / (x u y)

def jaccard(x,y):
  numerator = x.intersection(y)
  denomenator = x.union(y)  #this nume and deno things give character set
  return len(numerator) / len(denomenator)   #we cannot divide character set by another character set,
                                    #we have to get the len of ot
```

**3)Implement the SOUNDEX algorithm**

```python
def soundeX(name):

  #name = "Herman"
  name  = name.upper()  #set all the letter to upper case

  soundeX =""

  #1st rule - we have to give first letter to soundeX
  soundeX += name[0]

  #2nd and 3rd rules
  dictionary = {"BFPV":"1", "CGJKQSXZ":"2", "DT":"3", "L":"4", "MN":"5", "R":"6", "AEIOUHWY":"0"}

  for char in name[1:]:
    for key in dictionary.keys():
      if char in key:  #that char is in that dictioanry
        digit = dictionary[key]
        #print(digit)

        #4th rule - remove all the consecutive pairs
        if digit != soundeX[-1]:
          soundeX +=digit
          #5th rule
          soundeX = soundeX.replace("0","")
          #6th rule - finally soundeX should contain only 4 characters
          soundeX=soundeX[:4].ljust(4,"0")
  print(soundeX)

```
