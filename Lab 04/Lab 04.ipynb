```python
from sklearn.feature_extraction.text import TfidfVectorizer    #to solve vectorize
from sklearn.metrics.pairwise import cosine_similarity      #to fine  cosine similarity
```

**Cosine similarity means the similarty between document and the word**

**We have to find tf-idf in each and every word in the document**

**After you can get arrays, that's consider as the vectors, By using those vectors you can find the cosine similarity between those vectors**

```python
document1 = 'the man went out for a walk'
document2 =  'the children sat around the fire'
```

```python
#make the document list
docs = [document1,document2]
```

# **IF-IDF**

---




```python
#to use TfidfVectorizer
vecs = TfidfVectorizer()
```

```python
#to vectorize those to documents
tfidf_mat = vecs.fit_transform(docs)
tfidf_mat
```

tf-idf verctorizer always remove "a"

```python
tfidf_mat.toarray()
```

First two rows for document 1 and last 2 for document 2, if-idf vectorizer automatically remove the "a"

After removing "a" , there are 10 terms

There are two elemenst in an array,
tfidf_mat[o] for document 1 vectors, tfidf_mat[1] for document 2 vectors

0 means, there are no word represent

# **Cosine Similarity**

---






```python
cs = cosine_similarity(tfidf_mat[0],tfidf_mat[1])
cs
```
